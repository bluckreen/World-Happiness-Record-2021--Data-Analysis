# WORLD HAPPINESS ANALYSIS 2021

This project focuses on analyzing global happiness data to understand what factors contribute to a country's overall well-being. The dataset includes indicators such as GDP per capita, social support, healthy life expectancy, freedom to make life choices, generosity, and perceptions of corruption.

The main objective is to study how these variables influence the happiness score of different countries and to identify which factors have the strongest impact. Using basic data analysis and visualization techniques, the project compares happiness levels across countries and regions, highlights trends, and uncovers patterns in the data.

This project also aims to build an understanding of data handling, visualization, and interpretation, making it useful for developing foundational skills in data analytics.
